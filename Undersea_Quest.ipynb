```python
class Player():
    def __init__(self,name,oxygen=60):
        self.name = name
        self.oxygen = oxygen
        self.inventory = []
    def inventory_stat(self):
        if len(self.inventory)>0:
            print('Items in your inventory:')
            for items in self.inventory:
                print(items)
        else:
            print("You have nothing in your inventory.")
    def o2_stat(self):
        print(f"Current oxygen level: {round((self.oxygen/60)*100)}%")
    def o2_deplete(self,meters):
        self.oxygen = self.oxygen-meters
    def o2_add(self):
        self.oxygen = 60
    def add_inventory(self,item):
        self.inventory.append(item)
        print(f'{item} has been added to your inventory!')
    def remove_inventory(self,item):
        self.inventory.remove(item)
```

```python
def land_story(player,name,on_land_part):
    if on_land_part == 1:
        questions = []
        input('Let the adventure begin!')
        input("You open your eyes and find yourself laying in scratchy sand. As you sit up, you smell the briny air and hear the sound of\nwaves crashing all around you.\n")
        input("'Oh, good! You're awake!' A voice exclaims behind you. You turn and see your older brother, John, limping toward you from the\nshore.\n")
        while 'A' not in questions or 'B' not in questions or 'C' not in questions:
            answer = input("What would you like to say?\na) 'How did we get here?'\nb) 'Why are we here?'\nc) 'Why are you limping?'\n\n")
            if answer.upper() == 'A':
                input(f"\n'Wow {name}, you really must've hit your head hard. We were heading for this island on our boat. While we were bringing her\nonto shore a big wave came up and threw you into the rocks. You've been out cold since.'")
            elif answer.upper() == 'B':
                input(f"\n'{name}, how far back have you forgotten?? You remember mom, right? And how she's been sick with the Chill? Well, it just so\nhappens that the only known cure for the Chill, Purple Kelp, grows somewhere in the waters surrounding this island. We came to gather\nsome for her.'\n")
            elif answer.upper() == 'C':
                input("\n'Ah, I was exploring over on the farside of the shoreline and stepped on a sharp shell. It cut me really deep.'\n")
                input(f"'And uhh, about that... My foot just won't stop bleeding. I think I'll be alright, but I don't think it's safe for either of us if I go into the water with an open wound. All those sharks will have a feast. You're gonna have to get that Purple Kelp on\nyour own {name}...'\n")
            questions.append(answer.upper())
        on_land_part = 2
    if on_land_part == 2:
        answer = input("What would you like to do next?\na) Assess your surroundings\nb) Dive into the water\n\n")
        if answer.upper() == 'A':
            on_land_part = 3
            print("You stare out at the island around you. It's small, probably less than a quarter of a mile from shore to shore. On the closest shore to your right, is a pile of boulders with a rather large crab sitting on top looking directly at you.")
        elif answer.upper() == 'B':
            return on_land_part
    if on_land_part == 3:
        answer = input("What would you like to do next?\na) Approach the crab\nb) Dive into the water\n\n")
        if answer.upper() == 'A':
            on_land_part = 4
            input("You approach the crab warily. Something about it makes you uncomfortable. This doesn't seem like any ordinary crab.")
            input("Crab:\n'Hello there, human.'\n'If you'll pardon my eavesdropping, I couldn't help but overhear you and your friend talking. If it's Purple Kelp you're looking for, I can show you exactly where it grows.'")
            input("You follow the strange, polite crab who leads you to another larger pile of boulders downshore.")
            input("Crab:\n'If you dive from the very top rock, the Purple Kelp grows about 250 meters below the surface.'\n'But do be careful and make sure to monitor your oxygen levels. I hear that drowning is a horrendous fate...'")
        elif answer.upper() == 'B':
            return on_land_part
    while on_land_part == 4:
        answer = input("What would you like to do next?\na) Approach John\nb) Dive into the water\n\n")
        if answer.upper() == 'A' and len(player.inventory) > 0:
            input(f"John:\n'Well what are you waiting for {name}? Mother isn't getting any better. Go find that Purple Kelp!'\n")
        elif answer.upper() == 'A' and 'Sharp Shell' not in player.inventory:
            input("John:\n'So I was thinking about how you're going in there alone and I think I'd feel a lot better if you had something to protect yourself.'\n'I went back and found that shell I stepped on earlier. You should take it, just in case.'\n'Careful though, it's pretty sharp.'\n")
            player.add_inventory('Sharp Shell')
        elif answer.upper() == 'B':
            return on_land_part
```
